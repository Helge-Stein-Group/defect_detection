# Plot the defects on horizontal axis

```python
#import neessary libraries
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use(['science', 'nature', 'no-latex'])
plt.rcParams["figure.figsize"] = (5,3)

def plot_horizontal_defects(df_horizontal):
    #Extraxt the horizontal coordinates
    horizontal_coords=df_horizontal[['x0', 'x2']].to_numpy()
    horizontal_coords_copy=horizontal_coords.copy() # make a copy of the coordinates 
    horizontal_coords_copy = horizontal_coords_copy.ravel() 
    horizontal_coords_copy.sort() #sort the values in ascending order

    

    sorted_values_horizotal = [] #Create a empty list to sort the horizontal coordinates (horizontal_coords_copy) in ascending order
    for i in range(len(horizontal_coords_copy)-1):
        sorted_values_horizotal.append([horizontal_coords_copy[i], horizontal_coords_copy[i+1]])
        
        
    count_x=[] #Count the number of defects on a particular position
    for i in range(len(sorted_values_horizotal)):
        sum_x=0
        for j in range(len(horizontal_coords)):
            if ((horizontal_coords[j][0]<=sorted_values_horizotal[i][0]) & (horizontal_coords[j][1]>=sorted_values_horizotal[i][1])):
                sum_x = sum_x+1;
        count_x.append(sum_x)       
        
    # Pass the x and y cordinates of the bars
    csfont = {'fontname':'sans-serif'}    
    plt.bar(horizontal_coords_copy[:-1],count_x, width=[sorted_values_horizotal[i][1]-sorted_values_horizotal[i][0] for i in range(len(sorted_values_horizotal))],align='edge',alpha=0.7)
    plt.savefig('..\Figures\Yolo_defect_bar\All defects on horizontal axis.png',dpi=600)
    plt.tick_params(axis='both',direction='in',labelsize=8,size=4)
    plt.xlabel('pixel values on x axis', size=10, **csfont)
    plt.ylabel('Number of defects', size=10, **csfont)
    plt.show()

#Read the bounding box coordinates
df_bounding_box=pd.read_csv(r'..\yolov5\file1.csv',index_col=None)
plot_horizontal_defects(df_bounding_box)
```

#  Plot the defects on vertical axis
