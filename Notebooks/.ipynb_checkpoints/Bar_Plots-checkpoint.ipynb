# Plot the defects on horizontal axis

```python
#import neessary libraries
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('science')
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)

def plot_horizontal_defects(df_horizontal):
    #Extraxt the horizontal coordinates
    horizontal_coords=df_horizontal[['x0', 'x2']].to_numpy()
    h_array=horizontal_coords.copy() # make a copy of the coordinates 
    h_array = h_array.ravel() 
    h_array.sort() #sort the values in the array in ascending order


    new_arr_x = [] #Create an empty array to sort the array entries in ascending order
    for i in range(len(h_array)-1):
        new_arr_x.append([h_array[i], h_array[i+1]])

    count_x=[] #Count the number of defects on a particular position
    for i in range(len(new_arr_x)):
        sum_x=0
        for j in range(len(horizontal_coords)):
            if ((horizontal_coords[j][0]<=new_arr_x[i][0]) & (horizontal_coords[j][1]>=new_arr_x[i][1])):
                sum_x = sum_x+1;
        count_x.append(sum_x)


    # Pass the x and y cordinates of the bars to the
    plt.bar(h_array[:-1],count_x, width=[new_arr_x[i][1]-new_arr_x[i][0] for i in range(len(new_arr_x))],align='edge',alpha=0.4,color='green')
    plt.tick_params(axis='both',direction='in')
    #plt.legend()
    # The following commands add labels to our figure.
    plt.xlabel('pixel values on x axis')
    plt.ylabel('defects')
    plt.title('Horizontal defects Bar chart')
    plt.savefig('E:\yolov5_defect_detection\Results\Yolo_defect_bar\All defects on horizontal axis.png')
    plt.show()

#Read the bounding box coordinates
root = Path('..','yolov5', 'file1.csv') #dir where images and annotations stored
assert root.is_dir() # make sure this directory is found

df_bounding_box=pd.read_csv(r'E:\yolov5_defect_detection\yolov5\file1.csv',index_col=None)
plot_horizontal_defects(df_bounding_box)
```

#  Plot the defects on vertical axis

```python
def plot_vertical_defects(df_vertical):
    vertical_coords=df_vertical[['x1', 'x3']].to_numpy()
    v_array=vertical_coords.copy()
    v_array = v_array.ravel()
    v_array.sort()

    new_arr_y = []
    for i in range(len(v_array)-1):
        new_arr_y.append([v_array[i], v_array[i+1]])

    count_y=[]
    for i in range(len(new_arr_y)):
        sum_y=0
        for j in range(len(vertical_coords)):
            if ((vertical_coords[j][0]<=new_arr_y[i][0]) & (vertical_coords[j][1]>=new_arr_y[i][1])):
                sum_y = sum_y+1;
        count_y.append(sum_y)

    # Pass the x and y cordinates of the bars to the
    # function. The label argument gives a label to the data.
    plt.ylim([500,0])
    plt.tick_params(axis='both',direction='in')
    plt.barh(v_array[:-1],width=count_y, height=[new_arr_y[i][1]-new_arr_y[i][0] for i in range(len(new_arr_y))],align='edge',alpha=0.4,color='green')
    #plt.legend()
    # The following commands add labels to our figure.
    plt.xlabel('defects')
    plt.ylabel('pixel values on y axis')
    plt.title('Vertical defects Bar chart')

    plt.savefig('E:\yolov5_defect_detection\Results\Yolo_defect_bar\All defects on vertical axis.png')
    plt.show()
    
#Read the bounding box coordinates
df_bounding_box=pd.read_csv(r'E:\yolov5_defect_detection\yolov5\file1.csv',index_col=None)
plot_vertical_defects(df_bounding_box)
```

```python
#import neessary libraries
import pandas as pd
import matplotlib.pyplot as plt


def vertical_category_defects(df_all):
    grouped=df_all.groupby('class')

    for name,group in grouped:
        if name==0:
            #Extraxt the horizontal coordinates
            horizontal_coords_0=group[['x1', 'x3']].to_numpy()
            h_array_0=horizontal_coords_0.copy() # make a copy of the coordinates 
            h_array_0 = h_array_0.ravel() 
            h_array_0.sort() #sort the values in the array in ascending order


            new_arr_0 = [] #Create an empty array to sort the array entries in ascending order
            for i in range(len(h_array_0)-1):
                new_arr_0.append([h_array_0[i], h_array_0[i+1]])

            count_0=[] #Count the number of defects on a particular position
            for i in range(len(new_arr_0)):
                sum_0=0
                for j in range(len(horizontal_coords_0)):
                    if ((horizontal_coords_0[j][0]<=new_arr_0[i][0]) & (horizontal_coords_0[j][1]>=new_arr_0[i][1])):
                        sum_0 = sum_0+1;
                count_0.append(sum_0)


            figure, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, constrained_layout=True)
            #plt.subplots(constrained_layout=True)
            ax1.tick_params(axis='both',direction='in')
            ax1.set_xlabel('defects')
            ax1.set_ylabel('pixel values on y axis')
            ax1.set_title('Scratches')
            ax1.barh(h_array_0[:-1],width=count_0, height=[new_arr_0[i][1]-new_arr_0[i][0] for i in range(len(new_arr_0))], color='green',alpha=0.5,align='edge',label='Scr')


        elif name==1:
                #Extraxt the horizontal coordinates
                horizontal_coords_1=group[['x1', 'x3']].to_numpy()
                h_array_1=horizontal_coords_1.copy() # make a copy of the coordinates 
                h_array_1 = h_array_1.ravel() 
                h_array_1.sort() #sort the values in the array in ascending order


                new_arr_1 = [] #Create an empty array to sort the array entries in ascending order
                for i in range(len(h_array_1)-1):
                    new_arr_1.append([h_array_1[i], h_array_1[i+1]])

                count_1=[] #Count the number of defects on a particular position
                for i in range(len(new_arr_1)):
                    sum_1=0
                    for j in range(len(horizontal_coords_1)):
                        if ((horizontal_coords_1[j][0]<=new_arr_1[i][0]) & (horizontal_coords_1[j][1]>=new_arr_1[i][1])):
                            sum_1 = sum_1+1;
                    count_1.append(sum_1)


                ax2.tick_params(axis='both',direction='in')
                ax2.set_xlabel('defects')
                ax2.set_ylabel('pixel values on y axis')
                ax2.set_title('Bubbles')
                ax2.barh(h_array_1[:-1],width=count_1, height=[new_arr_1[i][1]-new_arr_1[i][0] for i in range(len(new_arr_1))],color='pink',alpha=0.4,align='edge',label='Bubb')
                #plt.legend()
                #The following commands add labels to our figure.
                #plt.xlabel('pixel values on x axis')
                #plt.ylabel('defects')
                #plt.title('Horizontal defects Bar chart')
                #plt.show()

        elif name==2:
            #Extraxt the horizontal coordinates
            horizontal_coords_2=group[['x1', 'x3']].to_numpy()
            h_array_2=horizontal_coords_2.copy() # make a copy of the coordinates 
            h_array_2 = h_array_2.ravel() 
            h_array_2.sort() #sort the values in the array in ascending order


            new_arr_2 = [] #Create an empty array to sort the array entries in ascending order
            for i in range(len(h_array_2)-1):
                new_arr_2.append([h_array_2[i], h_array_2[i+1]])

            count_2=[] #Count the number of defects on a particular position
            for i in range(len(new_arr_2)):
                sum_2=0
                for j in range(len(horizontal_coords_2)):
                    if ((horizontal_coords_2[j][0]<=new_arr_2[i][0]) & (horizontal_coords_2[j][1]>=new_arr_2[i][1])):
                        sum_2 = sum_2+1;
                count_2.append(sum_2)


            # Pass the x and y cordinates of the bars to the
            ax3.tick_params(axis='both',direction='in')
            ax3.set_xlabel('defects')
            ax3.set_ylabel('pixel values on y axis')
            ax3.set_title('Foil')
            ax3.barh(h_array_2[:-1],width=count_2, height=[new_arr_2[i][1]-new_arr_2[i][0] for i in range(len(new_arr_2))],alpha=0.5,align='edge', color='yellow',label='Foil')
            #plt.legend()
            # The following commands add labels to our figure.
            #plt.xlabel('pixel values on x axis')
            #plt.ylabel('defects')
            #plt.title('Horizontal defects Bar chart')
            #plt.show()

        else:
            #Extraxt the horizontal coordinates
            horizontal_coords_3=group[['x1', 'x3']].to_numpy()
            h_array_3=horizontal_coords_3.copy() # make a copy of the coordinates 
            h_array_3 = h_array_3.ravel() 
            h_array_3.sort() #sort the values in the array in ascending order


            new_arr_3 = [] #Create an empty array to sort the array entries in ascending order
            for i in range(len(h_array_3)-1):
                new_arr_3.append([h_array_3[i], h_array_3[i+1]])

            count_3=[] #Count the number of defects on a particular position
            for i in range(len(new_arr_3)):
                sum_3=0
                for j in range(len(horizontal_coords_3)):
                    if ((horizontal_coords_3[j][0]<=new_arr_3[i][0]) & (horizontal_coords_3[j][1]>=new_arr_3[i][1])):
                        sum_3 = sum_3+1;
                count_3.append(sum_3)


            # Pass the x and y cordinates of the bars to the
            ax4.tick_params(axis='both',direction='in')
            ax4.set_xlabel('defects')
            ax4.set_ylabel('pixel values on y axis')
            ax4.set_title('Agglomerates')
            ax4.barh(h_array_3[:-1],width=count_3, height=[new_arr_3[i][1]-new_arr_3[i][0] for i in range(len(new_arr_3))], color='red',alpha=0.4,align='edge',label='Agg')
            # The following commands add labels to our figure.

    plt.savefig('E:\yolov5_defect_detection\Results\Yolo_defect_bar\Subplots of defects on vertical axis.png')
    plt.show()


#Read the bounding box coordinates
df_bounding_box=pd.read_csv(r'E:\yolov5_defect_detection\yolov5\file1.csv',index_col=None)
vertical_category_defects(df_bounding_box)
```

# Defect plot of a single image

```python
###horizontal axis bars
import numpy as np
horizontal_coords = np.array([[ 426,  473],[ 149,  198],[1073, 1115],[1138, 1260]])
h_array=horizontal_coords.copy()
h_array = h_array.ravel()
h_array.sort()

new_arr_x = []
for i in range(len(h_array)-1):
    new_arr_x.append([h_array[i], h_array[i+1]])
    
count_x=[]
for i in range(len(new_arr_x)):
    sum_x=0
    for j in range(len(horizontal_coords)):
        if ((horizontal_coords[j][0]<=new_arr_x[i][0]) & (horizontal_coords[j][1]>=new_arr_x[i][1])):
            sum_x = sum_x+1;
    count_x.append(sum_x)

    
    
###vertical axis bars

vertical_coords = np.array([[475,496],[467,493],[34,101],[0,96]])
v_array=vertical_coords.copy()
v_array = v_array.ravel()
v_array.sort()

new_arr_y = []
for i in range(len(v_array)-1):
    new_arr_y.append([v_array[i], v_array[i+1]])
    
count_y=[]
for i in range(len(new_arr_y)):
    sum_y=0
    for j in range(len(vertical_coords)):
        if ((vertical_coords[j][0]<=new_arr_y[i][0]) & (vertical_coords[j][1]>=new_arr_y[i][1])):
            sum_y = sum_y+1;
    count_y.append(sum_y)

    
    
import pylab as plt
import matplotlib.image as mpimg

ax1 = plt.subplot2grid((2,2),(0, 0))

ax1.set_xlim([0, 1500])
ax1.set_ylim([500,0])
ax1.tick_params(direction="in")
ax1.set_title('Original Electrode Image')
img = mpimg.imread('D:\\YOLO_OBJ_DETECTION\\train_data\\images\\val\\Cognex_06-26-2020_14.06.53.jpg')
plt.xlim()
imgplot = plt.imshow(img)


ax2 = plt.subplot2grid((2,2),(0, 1))

ax2.set_ylim([500,0])
ax2.set_xlim([0,5])
ax2.tick_params(direction="in")
ax2.set_title('Defects on Vertical axis')
plt.barh(v_array[:-1],width=count_y, height=[new_arr_y[i][1]-new_arr_y[i][0] for i in range(len(new_arr_y))], align='edge',color='red',alpha=0.5)

#plt.show()

ax3 = plt.subplot2grid((2,2),(1, 0))
ax3.tick_params(direction="in")
ax3.set_xlim([0, 1500])
ax3.set_title('Defects on Horizontal axis')
plt.bar(h_array[:-1],count_x, width=[(new_arr_x[i][1]-new_arr_x[i][0]) for i in range(len(new_arr_x))], align='edge',color='red',alpha=0.5)

plt.tight_layout()
plt.savefig('Defects for a single image.png')
plt.show()
```
